## Introdução:
Este projeto foca na análise de dados de qualidade da água disponíveis no dataset "Water Quality" encontrado no site da Kaggle ("https://www.kaggle.com/datasets/adityakadiwal/water-potability"). Ao analisar esses dados, esperamos não apenas compreender as características que definem a água potável, mas também desenvolver um modelo de aprendizado de máquina supervisionado que possa prever a potabilidade da água com base nesses parâmetros. Isso permitirá intervenções mais rápidas e eficazes para garantir que a água consumida por populações esteja dentro dos padrões de segurança. Através deste estudo, pretendemos oferecer insights valiosos para o manejo de recursos hídricos e a formulação de políticas de saúde pública relacionadas à qualidade da água.

## Dicionário de dados:

* **pH**:O pH é um parâmetro importante na avaliação do equilíbrio ácido-base da água. É também o indicador da condição ácida ou alcalina do estado da água. A OMS recomendou o limite máximo permitido de pH de 6,5 a 8,5. Os actuais intervalos de investigação foram 6,52–6,83, que estão dentro dos padrões da OMS;

* **Dureza**: A dureza é causada principalmente por sais de cálcio e magnésio. Esses sais são dissolvidos em depósitos geológicos através dos quais a água viaja. O período de tempo que a água fica em contato com o material produtor de dureza ajuda a determinar quanta dureza existe na água bruta. A dureza foi originalmente definida como a capacidade da água de precipitar sabão causada pelo Cálcio e pelo Magnésio;

* **Solidos_totais_diss**: A água tem a capacidade de dissolver uma ampla gama de minerais ou sais inorgânicos e alguns orgânicos, como potássio, cálcio, sódio, bicarbonatos, cloretos, magnésio, sulfatos, etc. Esses minerais produziam sabor indesejado e cor diluída na aparência da água. Este é o parâmetro importante para o uso da água. A água com alto valor de TDS indica que a água é altamente mineralizada. O limite desejável para TDS é 500 mg/le o limite máximo é 1000 mg/l, prescrito para beber;

* **Cloraminas**: O cloro e a cloramina são os principais desinfetantes utilizados nos sistemas públicos de água. As cloraminas são mais comumente formadas quando a amônia é adicionada ao cloro para tratar a água potável. Níveis de cloro de até 4 miligramas por litro (mg/L ou 4 partes por milhão (ppm)) são considerados seguros na água potável;

* **Sulfatos**: Os sulfatos são substâncias naturais encontradas em minerais, solo e rochas. Eles estão presentes no ar ambiente, nas águas subterrâneas, nas plantas e nos alimentos. O principal uso comercial do sulfato é na indústria química. A concentração de sulfato na água do mar é de cerca de 2.700 miligramas por litro (mg/L). Varia de 3 a 30 mg/L na maioria das fontes de água doce, embora concentrações muito mais altas (1000 mg/L) sejam encontradas em algumas localizações geográficas;

* **Conditividade**: A água pura não é um bom condutor de corrente elétrica, mas sim um bom isolante. O aumento na concentração de íons aumenta a condutividade elétrica da água. Geralmente, a quantidade de sólidos dissolvidos na água determina a condutividade elétrica. A condutividade elétrica (CE), na verdade, mede o processo iônico de uma solução que lhe permite transmitir corrente. De acordo com os padrões da OMS, o valor CE não deve exceder 400 μS/cm;

* **Carbono_org**: O carbono orgânico total (COT) nas águas de origem provém da matéria orgânica natural em decomposição (MON), bem como de fontes sintéticas. TOC é uma medida da quantidade total de carbono em compostos orgânicos em água pura. De acordo com a EPA dos EUA, < 2 mg/L como TOC em água tratada/potável e < 4 mg/Lit em água de origem usada para tratamento;

* **Trihalometanos**: THMs são produtos químicos que podem ser encontrados na água tratada com cloro. A concentração de THMs na água potável varia de acordo com o nível de matéria orgânica na água, a quantidade de cloro necessária para tratar a água e a temperatura da água que está sendo tratada. Níveis de THM de até 80 ppm são considerados seguros na água potável;

* **Turbidez**: A turbidez da água depende da quantidade de matéria sólida presente no estado suspenso. É uma medida das propriedades de emissão de luz da água e o teste é usado para indicar a qualidade da descarga de resíduos em relação à matéria coloidal. O valor médio de turbidez obtido para Wondo Genet Campus (0,98 NTU) é inferior ao valor recomendado pela OMS de 5,00 NTU;

* **Potabilidade**: Indica se a água é segura para consumo humano, onde 1 significa Potável e 0 significa Não potável.;


## *Objetivo: Treinar o modelo de classificação para interpretar novos dados e identificar se um corpo dágua com os mesmos parametros analisados pelo modelo de enquadra em um corpo dágua com caracteristicas de um água própria ou impropria para o consumo humano.*


- Este é um problema de classificação para responder a uma pergunta específica sobre os dados e para isso trata-se de um modelo de aprendizado supervisionado. Com isso o problema central é desenvolver um modelo de aprendizado de máquina capaz de classificar a água em duas categorias: potável e não potável, facilitando assim a tomada de decisões críticas relacionadas ao tratamento da água, ao monitoramento ambiental e à saúde pública.
- Dado que o conjunto de dados de "Water Quality" no Kaggle é sinteticamente gerado e destinado apenas para fins de estudo e aprendizado, os dados podem não refletir totalmente as variações e complexidades dos dados reais, o que deve ser considerado na formulação das premissas e no desenvolvimento de hipóteses:
  - Representatividade Limitada: é importante reconhecer que podem não capturar todas as nuances e fatores ambientais reais que influenciam a potabilidade da água;
  - Independência das Observações: Continua-se a presumir que cada observação é independente, mesmo que os dados sejam sintéticos;
  - Erro e Ruído: Considera-se que os dados podem incluir um certo nível de erro ou ruído, assim como dados reais, especialmente porque são criados para fins de aprendizado e podem incluir variações intencionais para testar a robustez dos modelos desenvolvidos.

- Presume-se que o modelo desenvolvido poderá generalizar dentro do contexto de dados sintéticos, mas a aplicabilidade do modelo a dados reais pode necessitar de verificações e ajustes adicionais devido às simplificações e limitações do conjunto de dados sintético.

O processo de validação cruzada e testes em diferentes segmentos do conjunto de dados ajudará a entender melhor a robustez das hipóteses no contexto de dados sintéticos. Além disso, estudos futuros com dados reais podem ser necessários para validar a eficácia do modelo fora do ambiente de simulação.

## Divisão em Conjuntos de Treino e Teste

### Validação cruzada:
- Dado que o conjunto de dados é sinteticamente gerado, a validação cruzada pode não ser estritamente necessária para evitar viés de seleção do conjunto de dados.

### Divisão Simples em Treino e Teste:
- Uma abordagem simples de divisão em treino e teste pode ser suficiente, especialmente se o conjunto de dados for grande o suficiente e não houver restrições de recursos computacionais.

```python
X = df.drop('Potabilidade', axis=1).values

Y = df['Potabilidade'].values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=100)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

### Treinamento do Modelo.

```python
# Definição dos modelos
models = [
    LogisticRegression(max_iter=1000),
    RidgeClassifier(),
    SGDClassifier(),
    PassiveAggressiveClassifier(),
    Perceptron(),
    SVC(),
    LinearSVC(),
    NuSVC(),
    KNeighborsClassifier(),
    NearestCentroid(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    AdaBoostClassifier(),
    GradientBoostingClassifier(),
    GaussianNB(),
    BernoulliNB()
]

# Listas para armazenar os resultados e os nomes dos modelos
results = []
names = []

# Função para avaliar cada modelo
def evaluate_model(model, X_train, Y_train, X_test, Y_test):
    model.fit(X_train, Y_train)  # Treinar o modelo
    Y_pred = model.predict(X_test)  # Fazer previsões
    accuracy = accuracy_score(Y_test, Y_pred)  # Calcular a acurácia
    cv_scores = cross_val_score(model, X_train, Y_train, cv=10, scoring='accuracy')

    print(f"{model.__class__.__name__}:")
    print("Test Accuracy:", accuracy)
    print("CV Mean Accuracy:", cv_scores.mean())
    print("Classification Report:")
    print(classification_report(Y_test, Y_pred))
    print("-" * 60)

    return cv_scores  # Retorna os scores de validação cruzada para plotar

# Iterar sobre todos os modelos definidos
for model in models:
    cv_scores = evaluate_model(model, X_train, Y_train, X_test, Y_test)
    results.append(cv_scores)
    names.append(model.__class__.__name__)

# Boxplot de comparação dos modelos
fig = plt.figure(figsize=(14, 7))
fig.suptitle('Comparação da Acurácia dos Modelos com Cross Validation')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names, rotation=45, ha='right', fontsize=10)
plt.ylabel('Acurácia')
plt.show()
```

### Top 2 Modelos com Base na Acurácia Média da Validação Cruzada:

- RandomForestClassifier:
  - Test Accuracy: 0.673
  - CV Mean Accuracy: 0.664

- SVC:
  - Test Accuracy: 0.673
  - CV Mean Accuracy: 0.675

Ambos os modelos RandomForestClassifier e SVC tiveram uma acurácia média de validação cruzada em torno de 0.67, indicando uma boa capacidade de generalização para novos dados.

## Ajutes dos Hiperparametros

  - RandomForestClassifier:
    - n_estimators: vou experimentar aumentar o número padrão de estimadores para garantir uma boa cobertura do espaço de características, mas sem aumentar excessivamente o tempo de treinamento.
    - max_depth: vou definir um valor moderado para max_depth para evitar árvores muito profundas que se ajustem demais aos dados de treinamento.



```python
# RandomForestClassifier
rf_params = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

rf_grid = GridSearchCV(RandomForestClassifier(random_state=100), rf_params, cv=5, scoring='accuracy')
rf_grid.fit(X_train, Y_train)

print("Melhores Parâmetros para RandomForestClassifier:", rf_grid.best_params_)

# Treinamento do RandomForestClassifier com os melhores parâmetros
best_rf = RandomForestClassifier(**rf_grid.best_params_, random_state=100)
best_rf.fit(X_train, Y_train)


# SVC
svc_params = {
    'C': [0.1, 1, 10],
    'gamma': ['scale', 'auto']
}

svc_grid = GridSearchCV(SVC(random_state=100), svc_params, cv=5, scoring='accuracy')
svc_grid.fit(X_train, Y_train)

print("Melhores Parâmetros para SVC:", svc_grid.best_params_)

# Treinamento do SVC com os melhores parâmetros
best_svc = SVC(**svc_grid.best_params_, random_state=100)
best_svc.fit(X_train, Y_train)
```

## Usando o Ensemble para melhores resultados

```python
from sklearn.model_selection import RepeatedStratifiedKFold


# Definindo os modelos já treinados
rf_best_params = {'max_depth': 20, 'min_samples_split': 5, 'n_estimators': 100}
svc_best_params = {'C': 1, 'gamma': 'scale'}

rf_model = RandomForestClassifier(**rf_best_params, random_state=100)
svc_model = SVC(**svc_best_params, random_state=100, probability=True)

# Definindo o Voting Classifier com hard voting
voting_model = VotingClassifier(estimators=[
    ('RF', rf_model),
    ('SVC', svc_model)
], voting='hard')

# Definindo o scaler
scaler = StandardScaler()

# Definindo a validação cruzada
skf = RepeatedStratifiedKFold(n_splits=5, n_repeats=2)

accuracy = []

for train_index, test_index in skf.split(X, Y):
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = Y[train_index], Y[test_index]

    # Normalização dos dados
    scaler.fit(X_train)
    X_train_scaled = scaler.transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Treinamento do modelo de votação
    voting_model.fit(X_train_scaled, Y_train)

    # Predição
    predictions = voting_model.predict(X_test_scaled)

    # Avaliação do desempenho
    score = accuracy_score(Y_test, predictions)
    accuracy.append(score)

# Média da acurácia
mean_accuracy = np.mean(accuracy)
print("Acurácia Média do Voting Classifier (Hard Voting):", mean_accuracy)
```

## Analisando o desempenho dos modelos gerados em dados não vistos

```python
from sklearn.metrics import classification_report, confusion_matrix

# Modelo Voting Classifier
voting_predictions = voting_model.predict(X_test_scaled)
voting_accuracy = accuracy_score(Y_test, voting_predictions)
voting_classification_report = classification_report(Y_test, voting_predictions)
voting_confusion_matrix = confusion_matrix(Y_test, voting_predictions)

print("Voting Classifier (Hard Voting):")
print("Acurácia:", voting_accuracy)
print("Matriz de Confusão:")
print(voting_confusion_matrix)
print("Relatório de Classificação:")
print(voting_classification_report)
```

### Analise das métricas:
 - Acurácia (Accuracy):
    - 68.09% - o modelo alcançou uma acurácia de aproximadamente 68.09%, o que significa que cerca de 68.09% das previsões feitas pelo modelo estavam corretas. Isso indica um desempenho moderado do modelo em classificar corretamente as instâncias de água potável e não potável.

- Precisão (Precision):
    - Água Potável (Classe 0): 0.66
    - Não Potável (Classe 1): 0.81
    - Esses valores indicam uma capacidade razoável do modelo de fazer previsões precisas para ambas as classes.

- Revocação (Recall):
    - Água Potável (Classe 0): 0.96
    - Não Potável (Classe 1): 0.24
    - Isso sugere que o modelo tem uma tendência a subestimar as instâncias de água não potável, resultando em um recall baixo para essa classe.

- F1-score:
    - 0.79 (para água potável) e 0.37 (para água não potável)
    - Para a classe de água potável, o modelo obteve um F1-score de 0.79, indicando um bom equilíbrio entre precisão e recall. No entanto, para a classe de água não potável, o modelo obteve um F1-score de apenas 0.37, indicando um desequilíbrio entre precisão e recall para essa classe.

- Matriz de Confusão:
    - a matriz de confusão mostra que o modelo tem uma tendência a classificar corretamente instâncias de água potável (classe 0), mas tem dificuldades em identificar corretamente instâncias de água não potável (classe 1), resultando em um número considerável de falsos negativos para essa classe.
